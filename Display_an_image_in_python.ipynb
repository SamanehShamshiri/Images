# 1. USING CV2

```python
import cv2
import matplotlib.pyplot as plt
```

### a

```python
image_dir="puppy.jpg"
image=cv2.imread(image_dir)
plt.imshow(image)
plt.show()
```

### b

```python
image_dir="puppy.jpg"
image=cv2.imread(image_dir)
image=cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(image)
plt.show()
```

### c

```python
image_dir="puppy.jpg"
image=cv2.imread(image_dir)
image=cv2.imshow("puppy", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2.USING MATPLOTLIB

```python
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
```

```python
image=mpimg.imread("puppy.jpg")
plt.imshow(image)
plt.show()
```

### 3.USING PIL

```python
from PIL import Image
```

```python
image=Image.open("puppy.jpg")
image.show()
```

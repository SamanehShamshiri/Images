### 1. USING glob

```python
import glob
import cv2
import os
import matplotlib.pyplot as plt
```

```python
image_folder="figures"
print(os.listdir(image_folder))
```

```python
image_files=glob.glob("figures/*.jpg")
print(type(image_files))
print(len(image_files))                      
```

```python
for image in image_files[0:5]:
    img=cv2.imread(image)
    img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(image)
    plt.axis("off")
    plt.show()
```

### 2. USING os

```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
folder='figures'
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if img is not None:
            images.append(img)
    return images
```

```python
images_list=load_images_from_folder('figures')


for img in images_list[0:5]:
    plt.imshow(img)
    #plt.title(image)
    plt.axis("off")
    plt.show()
    
    
```
